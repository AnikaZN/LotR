```python
import pandas as pd
from pandas_profiling import ProfileReport
from sklearn.model_selection import train_test_split
import numpy as np


df = pd.read_csv('./lotr_clean.csv')
df.head()
```

```python
df['name']
```

```python
df['name_length'] = [len(x) for x in df['name']]
df['name_length']
```

```python
v = ['a', 'e', 'i', 'o', 'u', 'y', 'A', 'E', 'I', 'O', 'U', 'Y']

def vowel_count(column):
    all = []
    for name in column:
        chars = []
        for char in name:
            if char in v:
                chars.append(char)
        all.append(len(chars))
    return all

df['vowel_count'] = vowel_count(df['name'])
```

```python
df['pct_vowels'] = df['vowel_count'] / df['name_length']
df.head()
```

```python
df['first_letter'] = df['name'].str[0]
df.head()
```

```python
df['starts_vowel'] = df['first_letter'].isin(v)
df.head()
```

```python
df.to_csv('lotr_with_name_info.csv')
```
